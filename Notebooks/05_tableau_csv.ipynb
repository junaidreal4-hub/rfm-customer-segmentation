```python
# Prepare clean data file for Tableau
import pandas as pd

# Load your expanded RFM data
rfm_df = pd.read_csv("../Data/Processed/rfm_table_expanded.csv")

# Load segments (if you have it from earlier notebooks)
segments_df = pd.read_csv("../Data/Processed/rfm_scored_segments.csv")

# Merge RFM with segment labels
rfm_for_tableau = rfm_df.merge(
    segments_df[['CustomerID', 'Segment']], 
    on='CustomerID', 
    how='left'
)

# Save for Tableau
rfm_for_tableau.to_csv("../Data/Processed/tableau_data.csv", index=False)

print(" Tableau file created!")
print(f"Rows: {len(rfm_for_tableau):,}")
print(f"Columns: {list(rfm_for_tableau.columns)}")
```
